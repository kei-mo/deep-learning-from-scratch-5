## モンテカルロ積分


p(x)が一様分布の場合、積分は次のようになる。
- 1~Tまでのf(t)の和(積分)×1/Tは、一様分布u(t)に関するf(t)期待値と等しい。
- モンテカルロ法を使って関数の「積分」を「ランダムサンプルの平均」で近似できる


$$
\frac{1}{T}\sum_{t=1}^{T} f(t) 
= \, \mathbb{E}_{u(t)}\bigl[f(t)\bigr].

$$

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2

# サンプル数
T = 10000

# [0,1] の一様分布から T 個のサンプルを生成
X = np.random.rand(T)

# サンプルを f に通した値
fX = f(X)

# 部分和の平均を逐次計算
running_means = np.cumsum(fX) / (np.arange(T) + 1)

# 理論値 (解析解)
true_value = 1.0 / 3.0

plt.figure(figsize=(8,5))
plt.plot(running_means, label='Monte Carlo Estimate')
plt.axhline(true_value, color='red', linestyle='--', label='True Value = 1/3')
plt.xlabel('Number of Samples')
plt.ylabel('Estimate of Integral')
plt.xscale('log')
plt.title('Monte Carlo Integration for f(x) = x^2')
plt.legend()
plt.show()
```
